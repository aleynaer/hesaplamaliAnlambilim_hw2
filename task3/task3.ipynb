```python
addr = r"C:\Users\Asus\Desktop\Anlambilim_odev2\task3\texts\dataset\*txt"

corpus = [] 

import glob 
import errno 
 
files = glob.glob(addr) 
for name in files: 
    try: 
        with open(name) as f: 
            raw_text = f.read()
            corpus.append(raw_text)
            f.close()
    except IOError as exc: 
        if exc.errno != errno.EISDIR: 
            raise 
print(corpus)
```

```python
for i in corpus:
    i = " ".join(i.split())
```

```python
import pandas as pd
data = pd.DataFrame(corpus, columns = ["docs"])
data
```

```python
query_addr =  r"C:\Users\Asus\Desktop\Anlambilim_odev2\task3\texts\query\A11.txt"

with open(query_addr) as f: 
        query_text = f.read()
        f.close()
        
query_text
#%% add query to data
query_text = {"docs":query_text}
data = data.append(query_text, ignore_index=True)
data
```

```python
import fasttext
import fasttext.util
fasttext.util.download_model('en', if_exists='ignore')  # English
ft = fasttext.load_model('cc.en.300.bin')

#ft.get_dimension()
#300
fasttext.util.reduce_model(ft, 100)
ft.get_dimension()
```

```python
import numpy as np
a = np.zeros((100,))
data['ft_vectors'] = data.apply(lambda x: tuple(a), axis=1).apply(np.array)
data.head()
```

```python
data.head()
```

```python
for sample in range(len(data)):
    text = data["docs"].values[sample]
    text = " ".join(text.split())
    text = text.lstrip()
    text = text.lower()
    #print(text)
    a = np.zeros((100,)) # her metin için vektör tablosu oluştur
    counter = 0
    for w in text.split():
        word = w
        vector = ft.get_word_vector(word)  # metindeki kelimelerin uzaklık vektörünü çıkarır
        counter += 1
        #print(vector)
        a += vector # kelimelerin uzaklıklarını toplar
    a = a/counter # toplam uzaklığı kelime sayısına böler 
    #print(a)
    data["ft_vectors"].values[sample] = a # metin kelimelerin ortalama uzaklığı ile temsil edilir
```

```python
data
```

```python
query_vector = data["ft_vectors"].values[10]
query_vector
```

```python
def cos_sim(vector1, vector2):
    dot_product = np.dot(vector1, vector2)
    norm_1 = np.linalg.norm(vector1)
    norm_2 = np.linalg.norm(vector2)
    return dot_product / (norm_1 * norm_2)
```

```python
# fasttext ile sonuçlar
for i in range(len(data)-1):
    base_vector = data["ft_vectors"].values[i]
    similarity = cos_sim(base_vector, query_vector)
    print("doküman {0} ile sorgunun benzerliği: {1}".format(i+1,similarity))
```

## glove embeddings

```python
import numpy as np
a = np.zeros((100,))
data['glove_vectors'] = data.apply(lambda x: tuple(a), axis=1).apply(np.array)
data.head()
```

```python
embedding_dict = {}


with open(r'C:\Users\Asus\Desktop\Anlambilim_odev2\glove.6B.100d.txt', encoding = "utf8") as glove:
    for line in glove:
        values = line.split()
        word = values[0]
        vectors = np.asarray(values[1:], 'float32')
        embedding_dict[word] = vectors
        
glove.close()
```

```python
for sample in range(len(data)):
    text = data["docs"].values[sample]
    text = " ".join(text.split())
    text = text.lstrip()
    text = text.lower()
    #print(text)
    a = np.zeros((100,)) # her metin için vektör tablosu oluştur
    counter = 0
    for w in text.split():
        word = w
        vector = embedding_dict.get(word)  # metindeki kelimelerin uzaklık vektörünü çıkarır
        if(vector is None):
            vector = np.zeros((100,))
        counter += 1
        #print(vector)
        a += vector # kelimelerin uzaklıklarını toplar
    a = a/counter # toplam uzaklığı kelime sayısına böler 
    #print(a)
    data["glove_vectors"].values[sample] = a # metin kelimelerin ortalama uzaklığı ile temsil edilir
```

```python
data
```

```python
query_vector = data["glove_vectors"].values[10]
query_vector
```

```python
# glove ile sonuçlar
print("glove embeddings sonuçları")
for i in range(len(data)-1):
    base_vector = data["glove_vectors"].values[i]
    similarity = cos_sim(base_vector, query_vector)
    print("doküman {0} ile sorgunun benzerliği: {1}".format(i+1,similarity))
```
